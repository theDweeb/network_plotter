```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
with open('models.csv', 'w') as csvfile:
    filewriter = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    filewriter.writerow(['Model', 'Parameters', 'FLOPS', 'Top1', 'Top5'])
    filewriter.writerow(['MobileNetV2-1', '3.504', '300.79', '72.192', '90.534'])
    filewriter.writerow(['MobileNetV2-0.75', '2.636', '209.08', '69.952', '88.986'])
    filewriter.writerow(['MobileNetV2-0.50', '1.968', '97.14', '64.592', '85.392'])
    filewriter.writerow(['MobileNetV2-0.35', '1.677', '59.29', '60.092', '82.172'])
    
    filewriter.writerow(['EfficientNet-b0', '5.29', '390', '75.31', '92.3'])
    filewriter.writerow(['EfficientNet-bc1', '4.55', '310', '71.88', '90.53'])
    filewriter.writerow(['EfficientNet-bc2', '3.46', '180', '71.5', '90.30'])
    filewriter.writerow(['EfficientNet-bc3', '2.8', '110', '55', '77.366'])
```

```python
df = pd.read_csv('models.csv')
```

```python
params = df['Parameters']
flops = df['FLOPS']
top1 = df['Top1']
top5 = df['Top5']
efficientnets = df.iloc[-4:]
mobilenets = df.iloc[:4]
mobilenets
```

```python
x = efficientnets['Model']

mb1 = mobilenets['Parameters'].iloc[0]
mb1 = np.array([mb1 for i in range(len(x))])

mb2 = mobilenets['Parameters'].iloc[1]
mb2 = np.array([mb2 for i in range(len(x))])

mb3 = mobilenets['Parameters'].iloc[2]
mb3 = np.array([mb3 for i in range(len(x))])

mb4 = mobilenets['Parameters'].iloc[3]
mb4 = np.array([mb4 for i in range(len(x))])
```

```python

_, (p1) = plt.subplots(1, 1, figsize=(10,8));
p1.set_title("Model Parameters (M)");
p1.plot(x, efficientnets['Parameters'], 'k' , marker='o', label="EfficientNets");

p1.plot(x, mb1, '-r', label="MobileNetV2-1", marker='o');
p1.plot(x, mb2, '-g', label="MobileNetV2-0.75", marker='o');
p1.plot(x, mb3, '-b', label="MobileNetV2-0.50", marker='o');
p1.plot(x, mb4, '-m', label="MobileNetV2-0.35", marker='o');
# for var in (train, val):
#     p1.annotate('%0.2f' % var.min(), xy=(1, var.min()), xytext=(8, 0), 
#         xycoords=('axes fraction', 'data'), textcoords='offset points')
p1.set_xlabel("Models");
p1.set_ylabel("Params (M)");
p1.legend(loc='upper right');
```
